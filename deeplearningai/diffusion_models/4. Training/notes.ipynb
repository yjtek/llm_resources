## Training the NN

- Big picture
    - Take an observation from training data
    - Add noise to it
    - Feed into NN and ask it to predict that noise
    - The "loss" is computed by taking the difference between the predicted noise and the actual noise added to the image

- We want the NN to see as many versions of the image as possible going from most noisy to least noisy
    - So in training, we randomly sample a "timestep" that directly determines how much noise is added to the original observation
    - Results in more stable training
